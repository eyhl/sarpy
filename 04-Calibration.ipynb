```python
from sarpy.load import s1_load
import numpy as np
```

### load image

```python
path = "/home/simon/Desktop/dev_sarpy/test_data/S1A_IW_GRDH_1SDV_20181009T171427_20181009T171452_024062_02A131_E887.SAFE"
img = s1_load(path,['VV'],location=(43.20203134094203, 9.326176488289917),size=(4800,4000))
```

### Plot original VV

```python
img.simple_plot(0)
```

The image has an atribute which tells what kind of data it contains

```python
img.unit
```

### Calibrate the image

The mode option let you choose which calibration to perform. gamma is the recomended and the default.

```python
img_cal = img.calibrate(mode='gamma',tiles=3)
```

Plot of the calibrated image

```python
img_cal.simple_plot(0)
```

The unit of the new image is also updated

```python
img_cal.unit
```

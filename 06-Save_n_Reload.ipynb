```python
import sarpy
```

### Load and calibrate the image

```python
path = "/home/simon/Desktop/dev_sarpy/test_data/S1A_IW_GRDH_1SDV_20181009T171427_20181009T171452_024062_02A131_E887.SAFE"
img = sarpy.s1_load(path,location=(43.20203134094203, 9.326176488289917),size=(4800,4000))
img = img.calibrate('gamma',tiles=5)
```

### Plot the image

```python
img.simple_plot(0)
```

#### Show some additional information

```python
img.get_coordinate(1094,2031)
```

```python
img.time
```

```python
img.footprint
```

```python
img.save('test_save1')
```

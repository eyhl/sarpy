### Using sarpy (my own code)
import pakages and on code. The import works only if the notebook is located in the same folder as the sarpy folder (The inner one). Like it is downloaded 

```python
from sarpy.load import sar_load
import matplotlib.pyplot as plt
import numpy as np
```

Set the path to the unziped Image

```python
path = '/home/simon/Desktop/dev_sarpy/S1A_IW_GRDH_1SDV_20181009T171427_20181009T171452_024062_02A131_E887.SAFE'
```

check out the load function

```python
sar_load?
```

Load a subset of the image

```python
img = sar_load(path,location=(43.20203134094203, 9.326176488289917),size=(4800,4000))
```

See the band names

```python
img.band_names
```

Each band is saved as a numpy array aka a matrix. Note that index in python starts from 0. So to get the mean of the first band

```python
np.mean(img.bands[0])
```

Note. The image is not calibrated at all so this mean does not reflect anything real

The img object has a simple plot method

```python
img.simple_plot?
```

plot VV

```python
img.simple_plot(0)
```

plot VH

```python
img.simple_plot(1)
```

The object has a function to get coordinates from index or reversed

```python
img.get_coordinate(1000,2000)
```

```python
img.get_index(43.07674283633739, 9.358148722912384)
```

Note that the index is shifted a bit. This is a bug

It is possible to slice the image like an numpy array

```python
img_slice = img[2000:3800,2000:]
img_slice.simple_plot(0)
```

The object also contains the meta data

```python
img.product_meta
```
